```python
from selenium import webdriver
```

```python
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.google.com")
```

```python
search_bar = driver.find_element(By.NAME, "q")
search_bar.clear()
search_bar.send_keys("pycon")
search_bar.send_keys(Keys.RETURN)
print(driver.title)
```

##### webdriver → lets Python control a browser
##### Keys → allows keyboard actions (like pressing Enter)
#### Launches a new Chrome window controlled by the script
#### Navigates to the official Python website
#### Locates the search box using its name="q" attribute
#### Clears anything in the search field
#### Types the word "pycon"
#### Simulates hitting the Enter key to submit the search
