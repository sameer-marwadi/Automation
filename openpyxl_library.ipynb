```python
from openpyxl import Workbook, load_workbook
```

# Create a new Workbook

```python
wb = Workbook()
ws = wb.active
```

```python
ws['A1'] = 'Hello'
```

```python
ws['A2'] = 'World'
```

```python
wb.save('example.xlsx')
```

# Read an existing workbook

```python
wb = load_workbook('example.xlsx')
ws =wb.active
print(ws['A1'].value)
print(ws['A2'].value)
```

```python
for cell in ws["A"]:  # Column A
    print(cell.value)
```
