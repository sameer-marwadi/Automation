```python
import requests
```

```python
response = requests.get('https://api.github.com')
```

```python
print(f"Status code {response.status_code}")
```

```python
print(f"Content: {response.json()}")
```

#  response.json()

##### This calls the .json() method on a response object /
##### Typically, response comes from a library like requests /
